```python
from scipy import signal
import matplotlib.pyplot as plt
import numpy as np

fs = 200
N = fs*10
f1 = 0.2
f2 = 8.0
f3 = 50.0

time = np.arange(N) / fs
signal_1 = 2*np.sin(2*np.pi*f1*time)
signal_2 = np.sin(2*np.pi*f2*time)
signal_3 = np.sin(2*np.pi*f3*time) # 50 Hz "noise"
trend = np.linspace(0., 10., N)
x = signal_1 + signal_2 + signal_3
x_trend = x + trend

f, t, Sxx = signal.spectrogram(x, fs, window='hann', nfft=512)
f_trend, t_trend, Sxx_trend = signal.spectrogram(x_trend, fs, window='hann', nfft=512)

# Plots
fig, axes = plt.subplots(3, 2, sharex='col', figsize=(15, 8))

axes[0, 0].plot(time, x)
axes[0, 0].set_title("Signal without trend")

axes[1, 0].pcolormesh(t, f, Sxx)
axes[1, 0].set_ylabel('Frequency [Hz]')
axes[1, 0].set_ylim(0, 60)

axes[2, 0].pcolormesh(t, f, Sxx)
axes[2, 0].set_ylabel('Frequency [Hz]')
axes[2, 0].set_xlabel('Time [sec]')
axes[2, 0].set_ylim(0, 10)

axes[0, 1].plot(time, x_trend)
axes[0, 1].set_title("Signal with trend")

axes[1, 1].pcolormesh(t_trend, f_trend, Sxx_trend)
axes[1, 1].set_ylabel('Frequency [Hz]')
axes[1, 1].set_ylim(0, 60)

axes[2, 1].pcolormesh(t_trend, f_trend, Sxx_trend)
axes[2, 1].set_ylabel('Frequency [Hz]')
axes[2, 1].set_xlabel('Time [sec]')
axes[2, 1].set_ylim(0, 10)

plt.tight_layout()
plt.show()
```
